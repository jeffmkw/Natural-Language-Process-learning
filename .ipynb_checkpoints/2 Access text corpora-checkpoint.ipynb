```python
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
    print(round(num_chars/num_words), round(num_words/num_sents), round(num_words/num_vocab), fileid)
    
```

```python
from nltk.corpus import brown
news_text = brown.words(categories='news')
fdist = nltk.FreqDist(w.lower() for w in news_text)
modals = ['can', 'could', 'may', 'might', 'must', 'will', 'i']
for m in modals:
    print(m + ':', fdist[m], end=' ')
    #the end=' ' enables printing in the same line
```

```python
test = ['what', 'who', 'when', 'where', 'why']
for word in test:
    print(word,':',fdist[word], end=' ')
```

```python
cfd = nltk.ConditionalFreqDist(
          (genre, word)
          for genre in brown.categories()
          for word in brown.words(categories=genre))
sorted(cfd)
```

```python
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
```

```python
from nltk.corpus import inaugural
inaugural.fileids()
[fileid[:4] for fileid in inaugural.fileids()]
```

```python
%matplotlib inline
```

```python
cfd = nltk.ConditionalFreqDist(
          (target, fileid[:4])
          for fileid in inaugural.fileids()
          for w in inaugural.words(fileid)
          for target in ['america', 'citizen']
          if w.lower().startswith(target))
cfd.plot()
```

#my first word puzzle!

```python
puzzle_letters = nltk.FreqDist('egivrvonl')
obligatory = 'r'
wordlist = nltk.corpus.words.words()
```

```python
finallist = [w for w in wordlist
            if len(w) >= 6
            and obligatory in w
            #the FreqDist 
            and nltk.FreqDist(w) <= puzzle_letters]
finallist
```

```python
# now start comparing the names in different gender to see the overlapping
names = nltk.corpus.names
names.fileids()
```

```python
male_names = names.words('male.txt')
female_names = names.words('female.txt')
```

```python
[w for w in male_names if w in female_names]
```

```python
[w for w in male_names and w in female_names]
#turns out that have to start with a if
```

```python
# now to compare the different names ending with different letter.
cfd = nltk.ConditionalFreqDist(
          (fileid, name[-1])
          for fileid in names.fileids()
          for name in names.words(fileid))
cfd
```

```python
cfd.plot()
```
